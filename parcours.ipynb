```python
import pandas as pd
```

```python
parcours = pd.read_csv('parcours.csv').drop_duplicates()
parcours = parcours.reset_index().drop(columns='index')
parcours 
parcours['key'] = parcours['year'].astype('str') + '-' + parcours['day'].astype('str')
```

```python
parcours
```

```python
parcours[parcours['name'] == 'Parcours type:']['value'] = parcours[parcours['name'] == 'Parcours type:']['value'].apply(lambda x: x.replace('p',''))
```

```python
parcours_df = pd.pivot_table(parcours, values='value', index='key', columns='name', aggfunc='first').reset_index()
```

```python
parcours_sub = parcours_df[['key', 'Parcours type:', 'ProfileScore:', 'Race ranking:', 'Vert. meters:']]
parcours_sub['Parcours type:'] = parcours_sub['Parcours type:'].apply(lambda x: x.replace('p',''))
parcours_sub[['Parcours type:', 'ProfileScore:', 'Race ranking:', 'Vert. meters:']] = parcours_sub[['Parcours type:', 'ProfileScore:', 'Race ranking:', 'Vert. meters:']].astype('int')
parcours_sub
```

```python
parcours_sub['year'] = parcours_sub['key'].apply(lambda x: x.split('-')[0])
parcours_sub['day'] = parcours_sub['key'].apply(lambda x: x.split('-')[1])
```

```python
parcours_sub.to_csv('parcours_clean.csv')
```
